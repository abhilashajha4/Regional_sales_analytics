# Data Cleaning and Wrangling


```python
df = df_sales.merge(
    df_products,
    how='left',
    left_on='Customer Name Index',
    right_on='Customer Index'
)
```

```python
df.head(5)
```

```python
df = df.merge(
    df_customers,
    how='left',
    left_on='Product Description Index',
    right_on='Index'
)
```

```python
df = df.merge(
    df_regions,
    how='left',
    left_on='Delivery Region Index',
    right_on='id'
)
```
